## Store teams info and roster

```python
for i, row in teams_df.iterrows():
    team_id = row['team_id']
    team_name = row['name']
    
    # For each team, upload the player data (we assume that the team names are the keys)
    if team_name in team_players_dfs:
        
        players_df = team_players_dfs[team_name]
        players_df = players_df[players_df['player_id'].isin(keep_ids)]
        
        # We create the roster, separated by role
        goalkeepers = players_df[players_df['position'] == 'Goalkeeper'][['player_id', 'name']].to_dict(orient = 'records')
        defenders = players_df[players_df['position'] == 'Defender'][['player_id', 'name']].to_dict(orient = 'records')
        midfielders = players_df[players_df['position'] == 'Midfielder'][['player_id', 'name']].to_dict(orient = 'records')
        attackers = players_df[players_df['position'] == 'Attacker'][['player_id', 'name']].to_dict(orient = 'records')

        # Create doc for each team
        team_document = {
            "_id": team_id,
            "name": team_name,
            "corporate_information": {
                "foundation": row['founded'],
                "city": row['venue_city'],
                "stadium": row['venue_name'],
                "capacity": row['venue_capacity']
            },
            "roster": {
                "goalkeepers": goalkeepers,
                "defenders": defenders,
                "midfielders": midfielders,
                "attackers": attackers
            }
        }
        
        # Insert doc in the MongoDB collection
        teams_collection.insert_one(team_document)
```

## Store fanta stats

```python
# Load fanta data
grades_df = pd.read_csv('../1 - DataAcquisition/raw_data/df_players_grades.csv', index_col = 0)
grades_df
```

```python
# Creating a dictionary organised by matchdays and teams
documents = {}

for matchday in grades_df['matchday'].unique():
    matchday_data = grades_df[grades_df['matchday'] == matchday]
    
    # Dict for each matchday
    matchday_dict = {}
    
    for team in matchday_data['team_name'].unique():
        team_data = matchday_data[matchday_data['team_name'] == team]
        
        # List of players with their grades and stats
        players = []
        for _, row in team_data.iterrows():
            players.append({
                'Name': row['normalized_player_name'],
                 'Grade': row['grade'],
                 'FantaGrade': row['fantagrade'],
                 'GoalsScored': row['goals_scored'],
                 'GoalsConceded': row['goals_conceded'],
                 'Autogoals': row['autogoals'],
                 'PenaltiesScored': row['penalties_scored'],
                 'PenaltiesMissed': row['penalties_missed'],
                 'PenaltiesSaved': row['penalties_saved'],
                 'Assists': row['assists']  
            })
        
        # Add team with players to the matchday
        matchday_dict[team] = players
    
    # Add matchday to docs
    documents[int(matchday)] = matchday_dict

# Load all docs on MongoDB
for matchday, data in documents.items():
   votes_collection.insert_one({
        '_id': matchday,  # Use matchday as id of the doc
        'teams': data
   })
```

```python
selected_columns = [ 'team_name', 'appereances', 'lineups', 'minutes', 'sub_in', 'sub_out',
       'bench', 'total_shots', 'on_shots', 'scored_goals', 'conceded_goals',
       'saved_goals', 'assists', 'total_passes', 'key_passes',
       'passes_accuracy', 'tackles', 'blocks', 'interceptions', 'total_duels',
       'won_duels', 'attempted_dribbles', 'won_dribbles', 'drawn_fouls',
       'committed_fouls', 'yellow_cards', 'yellowred_cards', 'red_cards',
       'scored_penalties', 'missed_penalties', 'saved_penalties']
```

```python
#  Group data by player_id including only selected columns
grouped_stats = stats[selected_columns + ["player_id"]].groupby("player_id").apply(
    lambda x: x.to_dict(orient="records")
).to_dict()

# Aggiorna i documenti MongoDB
for player_id, stats in grouped_stats.items():
    players_collection.update_one(
        {"_id": player_id},  # Filtra per _id (player_id corrisponde a _id)
        {"$set": {"stats": stats}},  # Aggiungi il campo "stats"
        upsert=True  # Se il documento non esiste, lo crea
    )
```
